```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lakers2001 = pd.read_csv('2001lakers.csv')
```

```python
lakers_norooks2001 = lakers2001.dropna(axis=0)
```

```python
lakers_norooks2001_ws = lakers_norooks2001.drop(['2001 MP', '2000 MP', '2001 G', '2000 G', '2001 USG', '2000 USG', '2001 WS/48',
                                '2000 WS/48'], axis=1)
lakers_norooks2001_ws = lakers_norooks2001_ws.sort_values(by='2001 WS',ascending=False)
```

```python
lakers2001_norooks_2 = pd.melt(lakers_norooks2001_ws[:12], id_vars = "Name")
```

```python
lakers2001_norooks_2 = lakers2001_norooks_2.rename(columns={'Name':'Player', 'variable':'Season', 'value':'WS'})
```

```python
sns.catplot(x='Player', y='WS', hue='Season', data=lakers2001_norooks_2, kind='bar', palette=sns.color_palette(['purple', 'gold']), 
            height=8, aspect=2)
plt.title('Winshares for 2000-01 Lakers vs 1999-00 Lakers')
```

```python
lakers_norooks2001_usg = lakers_norooks2001.drop(['2001 MP', '2000 MP', '2001 G', '2000 G', '2001 WS', '2000 WS', '2001 WS/48',
                                '2000 WS/48'], axis=1)
```

```python
lakers_norooks2001_usg = lakers_norooks2001_usg.sort_values(by='2001 USG',ascending=False)
```

```python
lakers2001_norooks_usg = pd.melt(lakers_norooks2001_usg[:12], id_vars = "Name")
```

```python
lakers2001_norooks_usg = lakers2001_norooks_usg.rename(columns={'Name':'Player', 'variable':'Season', 'value':'USG'})
```

```python
sns.catplot(x='Player', y='USG', hue='Season', data=lakers2001_norooks_usg, kind='bar', palette=sns.color_palette(['purple', 'gold']), 
            height=8, aspect=2)
plt.title('Usage rates for 2000-01 Lakers vs 1999-00 Lakers')
```
