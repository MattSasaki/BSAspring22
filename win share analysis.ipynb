```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lakers2022 = pd.read_csv('2021lakers.csv')
```

```python
lakers2022
```

```python
lakers_norooks2022 = lakers2022.dropna(axis=0)
```

```python
lakers_norooks2022_ws = lakers_norooks2022.drop(['2021 MP', '2020 MP', '2021 G', '2020 G', '2021 USG', '2020 USG', '2021 WS/48',
                                '2020 WS/48'], axis=1)
lakers_norooks2022_ws = lakers_norooks2022_ws.sort_values(by='2021 WS',ascending=False)
```

```python
lakers2022_norooks_2 = pd.melt(lakers_norooks2022_ws[:12], id_vars = "Name")
```

```python
lakers2022_norooks_2 = lakers2022_norooks_2.rename(columns={'Name':'Player', 'variable':'Season', 'value':'WS'})
```

```python
sns.catplot(x='Player', y='WS', hue='Season', data=lakers2022_norooks_2, kind='bar', palette=sns.color_palette(['purple', 'gold']), 
            height=8, aspect=2)
plt.title('Winshares for 2021-22 Lakers vs 2020-21 Lakers')
```

```python
lakers_norooks2022_usg = lakers_norooks2022.drop(['2021 MP', '2020 MP', '2021 G', '2020 G', '2021 WS', '2020 WS', '2021 WS/48',
                                '2020 WS/48'], axis=1)
```

```python
lakers_norooks2022_usg = lakers_norooks2022_usg.sort_values(by='2021 USG',ascending=False)
```

```python
lakers2022_norooks_usg = pd.melt(lakers_norooks2022_usg[:12], id_vars = "Name")
```

```python
lakers2022_norooks_usg = lakers2022_norooks_usg.rename(columns={'Name':'Player', 'variable':'Season', 'value':'USG'})
```

```python
sns.catplot(x='Player', y='USG', hue='Season', data=lakers2022_norooks_usg, kind='bar', palette=sns.color_palette(['purple', 'gold']), 
            height=8, aspect=2)
plt.title('Usage rates for 2021-22 Lakers vs 2020-21 Lakers')
```
