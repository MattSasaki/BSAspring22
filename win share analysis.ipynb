```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lakers = pd.read_csv('2021lakers.csv')
```

```python
lakers
```

```python
lakers_norooks = lakers.dropna(axis=0)
```

```python
lakers_norooks_ws = lakers_norooks.drop(['2021 MP', '2020 MP', '2021 G', '2020 G', '2021 USG', '2020 USG', '2021 WS/48',
                                '2020 WS/48'], axis=1)
lakers_norooks_ws = lakers_norooks_ws.sort_values(by='2021 WS',ascending=False)
```

```python
norooks2 = pd.melt(lakers_norooks_ws[:12], id_vars = "Name")
```

```python
norooks2
```

```python
sns.catplot(x='Name', y='value', hue='variable', data=norooks2, kind='bar', palette=sns.color_palette(['purple', 'gold']), 
            height=8, aspect=2)
plt.title('Winshares for 2021-22 Lakers vs 2020-21 Lakers')
```

```python
lakers_norooks_usg = lakers_norooks.drop(['2021 MP', '2020 MP', '2021 G', '2020 G', '2021 WS', '2020 WS', '2021 WS/48',
                                '2020 WS/48'], axis=1)
```

```python
lakers_norooks_usg = lakers_norooks_usg.sort_values(by='2021 USG',ascending=False)
```

```python
norooks_usg = pd.melt(lakers_norooks_usg[:12], id_vars = "Name")
```

```python
norooks_usg
```

```python
sns.catplot(x='Name', y='value', hue='variable', data=norooks_usg, kind='bar', palette=sns.color_palette(['purple', 'gold']), 
            height=8, aspect=2)
plt.title('Usage rates for 2021-22 Lakers vs 2020-21 Lakers')
```
